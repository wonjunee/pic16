## Vertical Edge Detection

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating an image

image = np.zeros((100,100,3))
image[30:70,30:70]=1

plt.imshow(image)
plt.show()
```

```python
img_copy = image[:,:,0].copy()

# Create Filter for Vertical filter
fil = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])

print(fil)
```

```python
h,w = img_copy.shape
edge_im = np.zeros_like(img_copy)

plt.imshow(edge_im)
plt.show()
```

```python
for i in range(1,h-2):
    for j in range(1,w-2):
        edge_im[i,j] = np.sum(img_copy[i-1:i+2,j-1:j+2]*fil)

plt.imshow(edge_im)
plt.show()
```

```python
# Need to normalize it
edge_im_normalized = (edge_im+4.0)/8.0
plt.imshow(edge_im_normalized)
plt.show()
```

## Gaussian Blurring

```python
# Creating a Gaussian kernel

size = 5
sigma= 20
size = int(size) // 2
x, y = np.mgrid[-size:size+1, -size:size+1]
print(x)
print(y)
```

```python
fil =  np.exp(-((x**2 + y**2) / (2.0*sigma**2))) * 1.0 / (2.0 * np.pi * sigma**2)
fil = fil/np.sum(fil)
```

```python
image = np.zeros((100,100,3))
image[30:70,30:70]=1

img_copy = image[:,:,0].copy()
blurr_im = np.zeros_like(img_copy)

for i in range(2,h-3):
    for j in range(2,w-3):
        blurr_im[i,j] = np.sum(img_copy[i-2:i+3,j-2:j+3]*fil)

blurr_im3 = np.dstack((blurr_im,blurr_im,blurr_im))
plt.imshow(blurr_im3)
plt.show()
```
